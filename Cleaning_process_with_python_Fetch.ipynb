```python
df1.drop(columns=['CATEGORY_4'], inplace=True, errors='ignore')

cols_to_fill = ['CATEGORY_1', 'CATEGORY_2', 'CATEGORY_3', 'MANUFACTURER', 'BRAND']
df1[cols_to_fill] = df1[cols_to_fill].fillna('Unknown')

#is used to remove rows from a DataFrame (df1) where the BARCODE column has missing values (NaN).
df1.dropna(subset=['BARCODE'], inplace=True)

#convert a barcode from float to string because barcodes are identifiers, not numbers for calculations. Storing them as strings prevents issues like losing precision, removing leading zeros, or formatting errors.
df1['BARCODE'] = df1['BARCODE'].astype(str)

df1.drop_duplicates(inplace=True)

```

```python
for col in df1.columns:
    print(f"Value counts for {col}:\n{df1[col].value_counts()}\n")
```

```python
df1.to_csv('CLEANPRODUCTS_TAKEHOME.csv', index=False)

print("DataFrame saved as CLEANPRODUCTS_TAKEHOME.csv")
```

```python
# Convert dates to datetime
df2['PURCHASE_DATE'] = pd.to_datetime(df2['PURCHASE_DATE'], errors='coerce')
df2['SCAN_DATE'] = pd.to_datetime(df2['SCAN_DATE'], errors='coerce')

# Convert FINAL_QUANTITY and FINAL_SALE to numeric
df2['FINAL_QUANTITY'] = pd.to_numeric(df2['FINAL_QUANTITY'], errors='coerce')
df2['FINAL_SALE'] = pd.to_numeric(df2['FINAL_SALE'], errors='coerce')


#is used to remove rows from a DataFrame (df2) where the BARCODE column has missing values (NaN).
df2.dropna(subset=['BARCODE'], inplace=True)
```

```python
df2['FINAL_QUANTITY'] = df2['FINAL_QUANTITY'].fillna(0)
df2['FINAL_SALE'] = df2['FINAL_SALE'].fillna(0)
```

```python
df3['GENDER'] = df3['GENDER'].str.lower().replace({
    'female': 'Female',
    'male': 'Male',
    'transgender': 'Transgender',
    'non_binary': 'Non-Binary',
    'not_listed': 'Other',
    'unknown': 'Unknown',
    'prefer_not_to_say': 'Prefer Not to Say'
}).fillna('Other')

df3['GENDER'] = df3['GENDER'].replace({
    'unknown': 'Other',
    'not_specified': 'Other',
    'prefer not to say': 'Prefer Not to Say',
    'other': 'Other'  # Fixed syntax here
})

# Check updated value counts
print(df3['GENDER'].value_counts())
```
